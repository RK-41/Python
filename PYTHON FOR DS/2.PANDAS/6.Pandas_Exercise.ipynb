```python
import pandas as pd
import numpy as np
```

```python
# 1. Importing data
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PYTHON FOR DS/0.FILES TO BE USED/rainfall.csv')
df.head(3)
```

```python
# 2. Checking for missing values

df.info()
```

```python
# 3. Finding the district that gets highest annual rainfall

dist = df['ANNUAL'].max()
df.loc[df['ANNUAL']==dist]
```

```python
# 4. Displaying top 5 states that gets highest annual rainfall

df.sort_values('ANNUAL', ascending=0).head(5)
```

```python
# 5. Dropping columns

df.drop(columns=['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'], inplace=True)
df
```

```python
# 6. Pivot table

piv_tab = pd.pivot_table(df, index='STATE_UT_NAME')
piv_tab
```

```python
# 7. District count in each state

df.groupby('STATE_UT_NAME').count()['DISTRICT']
```
