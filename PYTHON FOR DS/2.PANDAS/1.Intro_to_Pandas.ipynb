# INTRODUCTION TO PANDAS

## Pandas is an open-source library for real world data analysis in python.
## It's built on top of Numpy.
## Using Pandas, data can be cleaned, transformed, manipulated, and analyzed.
## It's suited for diff kinds of data including tabular as in a SQL table or a Excel spreadsheet, time series data, observational or statistical datasets.

## The next step is to explore the data.
## Exploring the data helps to:

# Performing Operations on the Data

*   Concatenating operations

# WHY PANDAS?

## Pandas is one of the most popular data wrangling and analysis tools because it:


*   has the capability to load huge sizes of data easily
*   provides us with extremely streamlined forms of data representation
*   can handle heterogeneous data, has extensive set of data manipulation features and makes data flexible and customizable.
