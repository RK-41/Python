# MASKING OPERATION

## The Masking Operation replaces values where the condition is True.

## Syntax:

> DataFrame.mask(cond, other=nan, inplace=False, axis=None)

*   cond: Where 'cond' is True, replace with corresponding value from 'other', otherwise keep the original value.

*   other: Entries where 'cond' is True are replaced with corresponding value from 'other'.

*   inplace: Whether to perform the operation in place on the data.

*   axis: Alignment axis.







## Creating a DataFrame using a list of dictionaries:

```python
marks = [{'PR':54, 'DIP':74, 'BIC':22},
         {'PR':64, 'DIP':18, 'BIC':68}]

marks_df = pd.DataFrame(marks, index=['St1', 'St2'])
marks_df
```

## Replacing marks < 30 with 'Fail'

```python
marks_df.mask(marks_df<30, 'Fail')
# Masking isn't done inplace
```

```python
marks_df
```

# PRESERVING INDEX

## Pandas preserves the index and column labels in the output.
## For binary operations such as addition and multiplication, Pandas will automatically align indices when passing the objects to the functions.

## Creating a DataFrame using a list of dictionaries:

```python
import pandas as pd
import numpy as np

marks = [{'PR':54, 'DIP':74, 'BIC':22},
         {'PR':64, 'DIP':18, 'BIC':68}]

marks_df = pd.DataFrame(marks, index=['St1', 'St2'])
marks_df
```

## Encrypting Marks as sine of Marks

```python
# Using sin() fn from numpy
encr_marks = np.sin(marks_df)
encr_marks
```

### The encrypted marks are with same indices as the original marks. This is called INDEX PRESERVATION.

## Resetting Index

### In case of a requirement where the indes has to be restored to the default index, reset_index() fn must be used.
### It adds the existing index as a new column in the DataFrame.
### This can be done as follows:

```python
encr_marks.reset_index(inplace=True)
encr_marks
```

# APPLY METHOD

### The apply() method is used to apply a function along an axis of the DataFrame.
### Syntax:


> DataFrame.apply(func, asiz=0, result_type=None)

*   func: Function to apply to each col or row.
*   axis: Axis along which the fn is applied. 0-col and 1-row.


*   result_type: One out of 'expand', 'reduce', or 'broadcast'(results will be broadcst to the original shape of the DataFrame, the original index and columns will be retained).







## Creating a DataFrame using a list of dictionaries:

```python
import pandas as pd
import numpy as np

marks = [{'PR':54, 'DIP':74, 'BIC':22},
         {'PR':64, 'DIP':18, 'BIC':68}]

marks_df = pd.DataFrame(marks, index=['St1', 'St2'])
marks_df
```

## Finding minimum marks scored in each subject:

```python
marks_df.apply(np.min, axis=0)
```

## Finding total marks scored in each subject:

```python
marks_df.apply(np.sum, axis=0)
```

## Finding total marks scored by each student:

```python
marks_df.apply(np.sum, axis=1)
```

## Assigning each student average marks subject-wise:

```python
marks_df.apply(func=np.mean, axis=0, result_type='broadcast')
```

# GROUPING OPERATION

## Pandas supports a group by feature to group data for aggregate oeprations.
## Syntax:


> DataFrame.groupby(by = column_name, axis, sort)


## Creating a DataFrame of Cars Details

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PYTHON FOR DS/0.FILES TO BE USED/auto_mpg.csv')
df.head(3)
```

## Prob. St.: Find cars manufactured in each year.

## Prob. St.: Some senior engineers in XYZ custom cars want to understant the effect of model year and number of cylinders on horsepower.
## Sol.: One of the engineers suggests about checking the mean, minimum and maximum horsepower based on no. of cylinders and model year. For such requirement, the agg() func can be combined with groupby() func.

### Creating a DataFrame grouped on cylinders and model_year and finding mean, min, and max of horsepower:

```python
grouped_multiple = df.groupby(['cylinders','model_year']).agg({'horsepower': ['mean','min','max']})
```

### Naming columns in grouped DataFrame:

```python
grouped_multiple.columns = ['hp_mean', 'hp_min', 'hp_max']
```

### Resetting index:

```python
grouped_multiple = grouped_multiple.reset_index()
```

### Viewing head of resulting DataFrame:

```python
grouped_multiple.head()
```

## Prob. St.: The engineers at XYZ custom cars want to know about the relationship between model year and acceleration of cars.
## Sol.: For better understanding, the grouped results can be sorted based on average acceleration of cars built in each model year.

```python
df.groupby(['model_year']).mean().sort_values('acceleration', ascending=False)[['acceleration']]
```

> pd.concat( [df1, df2], sort = False)



## Prob. St.: Combine marks of students belonging to two different sections.
## Sol.: Use concatenation to combine marks.

### Creating DataFrames of marks of students belonging to two sections:

```python
import pandas as pd
marks_A = {'PR':[82,50,91],
           'DIP':[84,58,62],
           'BIC':[74,60,58]}

df_A = pd.DataFrame(marks_A, index=['Tom', 'Bruce', 'Anya'])

marks_B = {'PR':[82,50,91],
           'DIP':[84,58,62],
           'BIC':[74,60,58]}

df_B = pd.DataFrame(marks_B, index=['Anne', 'Leo', 'Robert'])
```

## In case of DataFrames having different columns, concat() function won't be as effective as needed as it would result in some null values because of column mismatch.

## To resolve the above condition, the merge() func can be used which joins two tables based on a key.
## Syntax:


> pd.merge(df1, df2, how='inner')



### In Pandas, the 'merge' keyword automatically performs the 'inner join'. For other types of join, the 'how' parameter must be specified.

### **NOTE:** Refer to the Pandas documentation for exploring more functionalities of merge() and read about functions like append().

# CROSS TABLE

## The crosstab() func computes a simple cross tabulation of two (or more) factors.
## By default, it computes a frequency table of the factors unless an array of values and an aggregation function are passed.

## Syntax:


> pd.crosstab(index, columns, values=None, rownames=None, colnames=None, aggfunc=None, margins=False, margins_name='ALL', dropna=True, nomalize=False)

*   index: array-like, Series, or list of arrays/Series. (Values to group by in the rows)

*   columns: array-like, Series, or list of arrays/Series. (Values to group by in the columns)

*   values: array-like, optional. (Array of values to aggregate according to the factors. Requires aggfunc to be specified)

*   rownames: sequence, default None. (If passed, must match no. of row array passed)

*   colnames: sequence, default None. (If passed, must match the no. or col array passed)

*   aggfunc: function, optional. (If specified, required values be specified as well)

*   margins: bool, default False. (Add row/col margnis (subtotals))

*   margins_name: str, default 'All'. (Name of the row/col that will contain the totals when margins is True)

*   dropna: bool, default True. (Don't include columns whose entries are all NaN)

*   normalize: bool, {'all', 'index', 'columns'} or {0,1}, default False. (Normalise by dividing all values by the sum of values)



## Prob. St.: Find frequency distribution of diff no. of cylinders across diff years.
