# OBJECT ORIENTED INTERFACE IN MATPLOTLIB

## We can also create the plotitng elements using object oriented approach.
## The use of an object-oriented approach is recommended as it gives more control over customization of the plots.


## Steps to create the plot using object-oriented approach:

*   Creating a figure.
*   Setting up the axes.

*   Creating a plot using the axes object.
*   Creating multiple plots using the same axes object.

*   Setting up the title, label, and legend for a plot.








## Creating a Plot

### Creating a sample data for plotting:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,16)
# 1 to 15

y = x**2
# sq. of x values

z = (-y/2)+40
# arithmetic operation applied to y values
```

### Step 1: Creating an empty canvas or figure to plot the values and visualize the plot:

```python
fig = plt.figure()
```

### Step 2: Setting the axes for the figure:
#### The axes object is the region where the data can be plotted. A figure can have n no. of axes object.

```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])
```

### Step 3: Plotting a line on the axes:

### Syntax:

> ax.plot(x,y,color,label)



```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])

# plotting the line
ax.plot(x,y,color="r",label='line1')
```

### Step 4: Plotting multiple lines on the axes:

```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes
ax = fig.add_axes([0,0,1,1])

# plotting the lines
ax.plot(x,y,color="r",label='line1')
ax.plot(x,z,color="g",label='line2')
```

### Step 5: Setting the title for the plot and the labels for the axes:

```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes
ax = fig.add_axes([0,0,1,1])

# plotting the lines
ax.plot(x,y,color="r",label='line1')
ax.plot(x,z,color="g",label='line2')

# setting title
ax.set_title('Title of the Plot', fontsize=16)

# setting X-label
ax.set_xlabel('Label for X-axis', fontsize=16)

# setting Y-label
ax.set_ylabel('Label for Y-axis', fontsize=16)
```

### Step 6: Adding a legend to the plot:

```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes
ax = fig.add_axes([0,0,1,1])

# plotting the lines
ax.plot(x,y,color="r",label='line1')
ax.plot(x,z,color="g",label='line2')

# setting title
ax.set_title('Title for the Plot', fontsize=16)

# setting X-label
ax.set_xlabel('Label for X-axis', fontsize=16)

# setting Y-label
ax.set_ylabel('Label for Y-axis', fontsize=16)

# adding legend
ax.legend()
```

# CREATING MULTIPLE AXES

## Plotting two lines in different axes:

```python
import matplotlib.pyplot as plt
import numpy as np

# creating sample data for plotting
x = np.arange(1,16)
y = x**2
z = (-y/2)+40

# creating an empty canvas/figure
fig = plt.figure()

# setting axes - ax1 and ax2 as [left, bottom, width, height]
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.15,0.40,0.40,0.45])

# plotting line1
ax1.plot(x,y,color="r",label='line1')

# setting title for plot 1
ax1.set_title('Title for Plot 1', fontsize=16)

# setting x-label for plot 1
ax1.set_xlabel('Plot 1 (X-axis)', fontsize=16)

# setting y-label for plot 1
ax1.set_ylabel('Plot 2 (Y-axis)', fontsize=16)

# adding legend for plot 1
ax1.legend()

# plotting line2
ax2.plot(x,z,color="g",label='line2')

# setting title for plot 2
ax2.set_title('Title for Plot 2', fontsize=14)

# setting x-label for plot 2
ax2.set_xlabel('Plot 2 (X-axis)', fontsize=14)

# setting y-label for plot 2
ax2.set_ylabel('Plot 2 (Y-axis)', fontsize=14)

# adding legend for plot 2
ax2.legend()
```

### Here, two plots are created using two different axes (ax1, ax2). Though is seems like plot 2 is embedded in plot 1 due to the placement of the axes, plot 1 and plot 2 are completely different plots and can obly be accessed through ax1 and ax2 respectively.

# SUBPLOTS

## The subplots method is used to create a common layout for multiple plots. 'm*n' no. of plots can be created using a subplot where m and n represent the np. of rows and columns respectively.

## Creating a Subplot that has 1 row and 2 columns:
## Syntax:

> plt.subplot(nrows, ncols)



```python
# importing packages
import matplotlib.pyplot as plt
import numpy as np

# creating sample data for plotting
x = np.arange(1,16)
y = x**2
z = (-y/2)+40

# initializing two axes for the subplot
fig, (ax1, ax2) = plt.subplots(1,2)

# plotting line1
ax1.plot(x,y,color="r",label='line1')
ax1.set_title('Title for Plot1', fontsize=15)
ax1.set_xlabel('Plot1 - X axis', fontsize=15)
ax1.set_ylabel('Plot1 - Y axis', fontsize=15)

# plotting line2
ax2.plot(x,z,color="g",label='line2')
ax2.set_title('Title for Plot2', fontsize=15)
ax2.set_xlabel('Plot2 - X axis', fontsize=15)
ax2.set_ylabel('Plot2 - Y axis', fontsize=15)
```

### A subplot is crested with two diff plots aligned in 1 row and 2 columns.
### In this plot, y-label of second plot is overlapping with the first plot. To avoid this, 'fig.tight_layout()' must be added.

```python
# importing packages
import matplotlib.pyplot as plt
import numpy as np

# creating sample data for plotting
x = np.arange(1,16)
y = x**2
z = (-y/2)+40

# initializing two axes for the subplot
fig, (ax1, ax2) = plt.subplots(1,2)

# plotting line1
ax1.plot(x,y,color="r",label='line1')
ax1.set_title('Title for Plot1', fontsize=15)
ax1.set_xlabel('Plot1 - X axis', fontsize=15)
ax1.set_ylabel('Plot1 - Y axis', fontsize=15)

# plotting line2
ax2.plot(x,z,color="g",label='line2')
ax2.set_title('Title for Plot2', fontsize=15)
ax2.set_xlabel('Plot2 - X axis', fontsize=15)
ax2.set_ylabel('Plot2 - Y axis', fontsize=15)

# to avoid overlapping
fig.tight_layout()
```
