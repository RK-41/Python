# SUBPLOTS

## The subplots method is used to create a common layout for multiple plots. 'm*n' no. of plots can be created using a subplot where m and n represent the np. of rows and columns respectively.

## Creating a Subplot that has 1 row and 2 columns:
## Syntax:

> plt.subplot(nrows, ncols)



```python
# importing packages
import matplotlib.pyplot as plt
import numpy as np

# creating sample data for plotting
x = np.arange(1,16)
y = x**2
z = (-y/2)+40

# initializing two axes for the subplot
fig, (ax1, ax2) = plt.subplots(1,2)

# plotting line1
ax1.plot(x,y,color="r",label='line1')
ax1.set_title('Title for Plot1', fontsize=15)
ax1.set_xlabel('Plot1 - X axis', fontsize=15)
ax1.set_ylabel('Plot1 - Y axis', fontsize=15)

# plotting line2
ax2.plot(x,z,color="g",label='line2')
ax2.set_title('Title for Plot2', fontsize=15)
ax2.set_xlabel('Plot2 - X axis', fontsize=15)
ax2.set_ylabel('Plot2 - Y axis', fontsize=15)
```

### A subplot is crested with two diff plots aligned in 1 row and 2 columns.
### In this plot, y-label of second plot is overlapping with the first plot. To avoid this, 'fig.tight_layout()' must be added.

```python
# importing packages
import matplotlib.pyplot as plt
import numpy as np

# creating sample data for plotting
x = np.arange(1,16)
y = x**2
z = (-y/2)+40

# initializing two axes for the subplot
fig, (ax1, ax2) = plt.subplots(1,2)

# plotting line1
ax1.plot(x,y,color="r",label='line1')
ax1.set_title('Title for Plot1', fontsize=15)
ax1.set_xlabel('Plot1 - X axis', fontsize=15)
ax1.set_ylabel('Plot1 - Y axis', fontsize=15)

# plotting line2
ax2.plot(x,z,color="g",label='line2')
ax2.set_title('Title for Plot2', fontsize=15)
ax2.set_xlabel('Plot2 - X axis', fontsize=15)
ax2.set_ylabel('Plot2 - Y axis', fontsize=15)

# to avoid overlapping
fig.tight_layout()
```
