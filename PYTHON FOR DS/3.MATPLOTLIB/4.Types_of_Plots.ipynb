# BOXPLOT

## A boxplot gives a good indication of distribution of data about the median. Boxplots are a standardized way of displaying the distribution of data based on the five-number summary ('minimum', first quartile(Q1), median, third quartile(Q3), and 'maximum').

image.png

## Prob. St.: Some customers of XYZ Custom Cars are interested in the mileage range of the cars that are restored by the company. They also want to compare the distribution of average mileage and city mileage (25% less than the average mileage).


## Sol.: First, let's plot the avg. mileage 'mpg' from the data using a boxplot.

## Syntax:

> ax.boxplot(data)

*   ax: represents axes


```python
# importing the required packages
import matplotlib.pyplot as plt

# creating an empty canvas/figure
fig = plt.figure(figsize=[6,6])

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])
ax.set_title('Distribution of mileage')

# plotting boxplot
ax.boxplot(df['mpg'])
fig.show()
```

### There's no data for city mileage, but city mileage is 25% less than the average mileage i.e., 'mpg'.


### Next is to process the data for city mileage:

```python
# processing data for city mileage
df['city_mileage'] = df['mpg']*0.75
df.head()
```

### A new column 'city_mileage' is created.

### Next, the distribution of the average mileage and city mileage has to be compared:

```python
# Comparing average mpg and city mpg using boxplot

# importing the required packages
import matplotlib.pyplot as plt

# creating a list containing 'mpg' and 'city_mileage'
mpg_list = [df['mpg'], df['city_mileage']]

# creating an empty canvas/figure
fig = plt.figure()

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])
ax.set_title('Distribution of Average MPG vs City MPG')
ax.set_ylabel('Mileage per gallon')

# plotting box plot
ax.boxplot(mpg_list, widths = 0.5)
fig.show()
```

# SCATTER PLOT

## A scatter plot uses dots or markers to represent a value on the axes. The scatter plot is one of the simplest plots which can accept both quantitative and qualitative values, with a wide variety of applications in primitive data analysis.

## Several meaningful insights can be drawn from a scatter plot. For example, identifying the type of correlation between variables before diving deeper into predictions.

image.png

## Prob. St.: The engineers at XYZ custom cars want to infer the correlation between horsepower and mileage.

## Sol.: First, let's plot a scatter plot for horsepower and mileage.

## Syntax:

> ax.scatter(x,y,marker)

*   ax: represents axes
*   x: data on horizontal axis

*   y: data on vertical axis
*   marker: shape of data points (eg, 'o' for circle, 's' for squares, etc.)







### Creating a DataFrame for the data set of cars:

```python
import pandas as pd

# creating a dataframe of data of cars
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PYTHON FOR DS/0.FILES TO BE USED/auto_mpg.csv')
```

### Visualizing correlation b/w horsepower and mileage:

```python
# importing the required packages
import matplotlib.pyplot as plt

# creating an empty canvas/figure
fig = plt.figure(figsize=[10,5])

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])

# plotting scatter plot
ax.scatter(df['horsepower'], df['mpg'])

ax.set_title('Scatter Plot for Horsepower and Mileage', fontsize=15)
ax.set_xlabel('Horsepower', fontsize=15)
ax.set_ylabel('Mileage per gallon', fontsize=15)

fig.show()
```

### From the above image, it can be concluded that mileage and horsepower are negatively correlated to some extent.

### Visualizing the correlation b/w mileage and horsepower based on the origin of cars:

```python
# importing the required packages
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
# creating an empty canvas/figure
fig = plt.figure(figsize=[12,5])

# adding axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])

ax.set_title('Scatter Plot of Horsepower and Mileage based on Origin of Cars', fontsize=15)
ax.set_xlabel('Horsepower', fontsize=15)
ax.set_ylabel('Mileage per Gallon', fontsize=15)

# assigning diff colors for each origin
origin_list = df['origin'].unique()
colors = cm.jet(np.linspace(0,1,len(origin_list)))

# plotting a scatter plot based on origin
for origin, color in zip(origin_list, colors):
  x = df[df['origin'] == origin]['horsepower']
  y = df[df['origin'] == origin]['mpg']
  plt.scatter(x, y, color = color, label=origin)

plt.legend()
```

### In the above Scatter plot, diff colors are assigned to diff origins.

### From the above plot, following are the insights:

1.   Which country's cars has the highest mileage?
2.   Which country's cars has the highest horsepower?

### Ans: USA.

# BAR CHART

## A bar chart is a graph with rectangular bars that usually compare different categories. Each bar represents a particular category. The length of the bar indicates the total no. of values or items in that category.

## The bar graphs can be plotted vertically or horizontally.

## Prob. St.: The engineers at XYZ Custom Cars want to know the no. of cars released in each year.

## Sol.: A bar graph can be created for this problem.

## Syntax:

> ax.scatter(x,height,width,bottom,align)

*   ax: represents axes
*   x: data on horizontal axis
*   height: height of the bar
*   weight: weight of the bar (default: 0.8)
*   bottom: Y coordinates of bar bases (default: 0)
*   align: alignment of the bars to X coordinates (default value: 'center')







### Creating a DataFrame for the data set of cars:

```python
import pandas as pd

# creating a dataframe of data of cars
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PYTHON FOR DS/0.FILES TO BE USED/auto_mpg.csv')
```

### Creating a grouped DataFrame according to model year:

```python
# importing the required packages
import matplotlib.pyplot as plt

# creating a grouped dataframe according to model year
grouped_df = df.groupby(['model_year']).count()[['name']]
grouped_df.reset_index(inplace=True)

# renaming columns of grouped_df
# # new_col = {'model_year':'model_year', 'name':'no_of_cars'}
# # grouped_df.rename(columns = new_col, inplace=True)
grouped_df.columns = ['model_year', 'no_of_cars']
display(grouped_df)

# creating an empty canvas/figure
fig = plt.figure(figsize=[7,5])

# setting axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('No. of Cars in each Year', fontsize=15)
ax.set_ylabel('No. of Cars', fontsize=12)
ax.set_xlabel('Model Year', fontsize=12)

# plotting bar graph
ax.bar(grouped_df['model_year'], grouped_df['no_of_cars'])
fig.show()
```

# HISTOGRAM

## A histogram also represents data as rectangular bars. Unilike the bar graph, it is used for continuous data. Each bar groups the no. into intervals (bins) and the height of the bar is based on the no. of values that fall into the corresponding intervals.

## A histogram is idealy suited to obtain the frequency distribution of a given data.

## Prob. St.: The engineers at XYZ Custom Cars want to identify the distribution of horsepower.

## Sol.: This can be done with help of a histogram.

## Syntax:

> ax.hist(x,bins)

*   ax: represents axes
*   x: input values - single array or seq. of arrays
*   bins: int or seq or str. If bin is integer, it defines the no. of equal-width bins in the range.





### Creating a Histogram for Horsepower:

```python
# importing required packages
import matplotlib.pyplot as plt

# creating an empty canvas/figure
fig = plt.figure(figsize=[8,6])

# adding axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Distribution of Horsepower', fontsize=15)
ax.set_ylabel('Frequency', fontsize=13)
ax.set_xlabel('Horsepower', fontsize=13)

# plotting histogram
ax.hist(df['horsepower'], bins=20)
fig.show()
```

### From above histogram, it can be observed that most of the cars have the horsepower value ranging b/w 70 and 110.

# PIE CHART

## A pie chart divides the entire dataset into distinct groups. The chart consists of a circle split into pies and each pie represents a group.

## The size of the pie is proportional to the no. of items in each group compared to others.

## The sum of the pies in a pie chart will always be 100%.

## Prob. St.: The management of XYZ Custom Cars has decided to open a new branch and is yet to decide the location. They would like to concentrate more on the 'origin' of the cars to make a better decision.

## Sol.: The details on origin of the cars and their numbers can be represented to the stakeholders visually for their easy understanding.

## Let's visualize the data using a pie chart:

## Syntax:

> ax.pie(x, labels)

*   ax: represents the axes.
*   x: wedge size. 1D array.
*   labels: seq. of strings providing the label for wedges.





### Creating a DataFrame for the data set of cars:

```python
import pandas as pd

# creating a dataframe of data of cars
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PYTHON FOR DS/0.FILES TO BE USED/auto_mpg.csv')
```

### Creating a Sub-dataframe to plot the pie chart:

```python
# sub-df
pie_df = pd.DataFrame()

# getting the count of 'origin' and assigning to the df
pie_df['Count'] = df['origin'].value_counts()

# re-arranging the index
pie_df = pie_df.reset_index()

# re-naming the column
pie_df.rename(columns={'index':'Country'}, inplace=True)

pie_df
```

### Plotting the Pie Chart:

```python
# importing required packages
import matplotlib.pyplot as plt

# creating an empty canvas/figure
fig = plt.figure()

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])

ax.set_title('Origin of the Cars')

# plotting pie chart
ax.pie(pie_df['Count'], labels=pie_df['Country'])
fig.show()
```

### A basic Pie chart has been plotted using the ax.pie() function in matplotlib.

### This plot could also be made visually appealing by customizing other parameters present in the ax.pie() function.

### Following are the parameters:

*   explode: To get an elevated view for the selected pie
*   colors: To customize the colors for the plot
*   autopct: To add the percentage of the distribution in the pie chart
*   shadow: To add shadow to the plot
*   startangle: To change the starting angle of the pie chart.



### Re-Plotting the Pie Chart with some Customizations:

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[8,6])
ax = fig.add_axes([0,0,1,1])
ax.set_title('Origin of the Cars')

# setting explode values
explode = [0.01, 0.03, 0.03]

# setting colors
colors = ['yellow', 'red', 'green']

ax.pie(pie_df['Count'],
       labels=pie_df['Country'],
       autopct='%0.1f%%',
       explode=explode,
       shadow=True,
       startangle=250,
       colors=colors)

fig.show()
```

### From the plot above, following are the insights:

1.   Which will be the most suitable region to start a new branch?

> Ans: USA.


2.   What is the percentage of cars produced from Europe?

> Ans: 17.6%.
