# OBJECT ORIENTED INTERFACE IN MATPLOTLIB

## We can also create the plotitng elements using object oriented approach.
## The use of an object-oriented approach is recommended as it gives more control over customization of the plots.


## Steps to create the plot using object-oriented approach:

*   Creating a figure.
*   Setting up the axes.

*   Creating a plot using the axes object.
*   Creating multiple plots using the same axes object.

*   Setting up the title, label, and legend for a plot.








## Creating a Plot

### Creating a sample data for plotting:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,16)
# 1 to 15

y = x**2
# sq. of x values

z = (-y/2)+40
# arithmetic operation applied to y values
```

### Step 1: Creating an empty canvas or figure to plot the values and visualize the plot:

```python
fig = plt.figure()
```

### Step 2: Setting the axes for the figure:
#### The axes object is the region where the data can be plotted. A figure can have n no. of axes object.

```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])
```

### Step 3: Plotting a line on the axes:

### Syntax:

> ax.plot(x,y,color,label)



```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])

# plotting the line
ax.plot(x,y,color="r",label='line1')
```

### Step 4: Plotting multiple lines on the axes:

```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes
ax = fig.add_axes([0,0,1,1])

# plotting the lines
ax.plot(x,y,color="r",label='line1')
ax.plot(x,z,color="g",label='line2')
```

### Step 5: Setting the title for the plot and the labels for the axes:

```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes
ax = fig.add_axes([0,0,1,1])

# plotting the lines
ax.plot(x,y,color="r",label='line1')
ax.plot(x,z,color="g",label='line2')

# setting title
ax.set_title('Title of the Plot', fontsize=16)

# setting X-label
ax.set_xlabel('Label for X-axis', fontsize=16)

# setting Y-label
ax.set_ylabel('Label for Y-axis', fontsize=16)
```

### Step 6: Adding a legend to the plot:

```python
# creating an empty canvas/figure
fig = plt.figure()

# setting axes
ax = fig.add_axes([0,0,1,1])

# plotting the lines
ax.plot(x,y,color="r",label='line1')
ax.plot(x,z,color="g",label='line2')

# setting title
ax.set_title('Title for the Plot', fontsize=16)

# setting X-label
ax.set_xlabel('Label for X-axis', fontsize=16)

# setting Y-label
ax.set_ylabel('Label for Y-axis', fontsize=16)

# adding legend
ax.legend()
```
