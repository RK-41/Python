# CREATING MULTIPLE AXES

## Plotting two lines in different axes:

```python
import matplotlib.pyplot as plt
import numpy as np

# creating sample data for plotting
x = np.arange(1,16)
y = x**2
z = (-y/2)+40

# creating an empty canvas/figure
fig = plt.figure()

# setting axes - ax1 and ax2 as [left, bottom, width, height]
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.15,0.40,0.40,0.45])

# plotting line1
ax1.plot(x,y,color="r",label='line1')

# setting title for plot 1
ax1.set_title('Title for Plot 1', fontsize=16)

# setting x-label for plot 1
ax1.set_xlabel('Plot 1 (X-axis)', fontsize=16)

# setting y-label for plot 1
ax1.set_ylabel('Plot 2 (Y-axis)', fontsize=16)

# adding legend for plot 1
ax1.legend()

# plotting line2
ax2.plot(x,z,color="g",label='line2')

# setting title for plot 2
ax2.set_title('Title for Plot 2', fontsize=14)

# setting x-label for plot 2
ax2.set_xlabel('Plot 2 (X-axis)', fontsize=14)

# setting y-label for plot 2
ax2.set_ylabel('Plot 2 (Y-axis)', fontsize=14)

# adding legend for plot 2
ax2.legend()
```

### Here, two plots are created using two different axes (ax1, ax2). Though is seems like plot 2 is embedded in plot 1 due to the placement of the axes, plot 1 and plot 2 are completely different plots and can obly be accessed through ax1 and ax2 respectively.
