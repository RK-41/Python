# INTRODUCTION TO MATPLOTLIB

## Matplotlib is one of the most basic and popular Python libraries used for data visualization. It was developed for imitating the plotting capabilities of MATLAB.

## matplotlib.pyplot is used for 2D graphics in Python programming. It can be used in python shell as well, scripts, web applications servers, and other graphical user interface toolkits.

## Matplotlib uses libraries such as NumPy as a base for underlying operations.

## Approaches to plotting in Matplotlib:



1.   MATLAB way of plotting using matplotlib.pyplot. It's simple to use.
2.   Object-Oriented way of plotting for more control and customization.



# IMPORTING MATPLOTLIB

## To make use of the functionalities that are present in Matplotlib library, the package must be imported into the environment.

```python
import matplotlib.pyplot as plt
```

### 'plt' is the std. alias used for pyplot module of Matplotlib.

## Creating a simple Plot using Matplotlib.pyplot:

### Syntax:


> plt.plot(x,y)

*   x: data on the horizontal axis.
*   y: data on the vertical axis.




```python
# importing the required packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creating two arrays
X = np.array([1,2,3,4,5])
Y = X**3
```

```python
# plotting values using matplotlib.pylot
plt.plot(X,Y)
```

## matplotlib.pyplot is the simplest way of plotting. It creates the default elements like Figure and Axes required and then plots the data.
## This approach can be used to plot different kinds of graphc like Line, Bar, Scatter plots, Histogram, etc.
