# **Line Chart**

## Line Chart is drawn by interconnecting all data points using straight line segments in sequential order.

## It is used to analyze historic variations and trends in data.

## The individual data points are chronologically connected to obtain the patterns and draw meaningful inferences.

## Prob. St.: XYZ Custom Cars is planning to generate a report comparing the Mileage, Horsepower and Weight of the cars manufactured in different year, starting from 1970 to 1982.

## Sol.: As a data analyst, a line chart can be created to visualize the relationship between mileage, horsepower and the weight of the cars manufactured in different years.


### Creating a DataFrame for the data set of cars:

```python
import pandas as pd

# creating a dataframe of data of cars
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PYTHON FOR DS/0.FILES TO BE USED/auto_mpg.csv')
```

### Creating a DataFrame containing the mean values of mileage, horsepower and the weight of the cars based on the model year:

```python
line_df = df.groupby(by = ['model_year'])

# creating a sub-dataframe for the line graph
line_df = line_df.mean()
```

### Plotting the Line Chart for the Scenario:

```python
# importing the required packages
import matplotlib.pyplot as plt

# creating an empty canvas/figure
fig = plt.figure(figsize=[8,4])

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])

# setting attributes
ax.set_title('Mileage vs Horsepower vs Weight', fontsize=16)
ax.set_xlabel('Model Year', fontsize=12)
ax.set_ylabel('Value', fontsize=12)

# plotting the lines
ax.plot(line_df['mpg']/10, label='Mileage /10')
ax.plot(line_df['horsepower']/100, label='Horsepower /100')
ax.plot(line_df['weight']/1000, label='Weight /1000')

ax.legend()
fig.show()
```

# Linestyle

## Matplotlib supports several line styles such as solid line, dashed line (----), dotted line (....), dashdot (-.-.-.-), etc.

## In addition to color and linestyle, the width of the line can be customized to get a unique plot.

### Customizing plots using linestyle, width and color:

```python
import matplotlib.pyplot as plt

# creating an empty canvas/fig
fig = plt.figure(figsize=[8,4])

# setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])

# plotting the lines
ax.plot(line_df['mpg']/10, label='Mileage', linestyle='-.', linewidth=2)
ax.plot(line_df['horsepower']/100, label='Horsepower', linestyle=':', linewidth=3)
ax.plot(line_df['weight']/1000, label='Weight', linestyle='--', linewidth=2)

# setting title, xlabel and ylabel
ax.set_title('Mileage vs Horsepower vs Weight', fontsize=15)
ax.set_xlabel('Model Year', fontsize=12)
ax.set_ylabel('Values', fontsize=12)

ax.legend()

fig.show()
```

# Text and Annotate

## In addition to adding the title and axis names, text can also be added to describe a plot.

### Adding text for a plot using the 'plt.text()' function:

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[8,4])
ax = fig.add_axes([0,0,1,1])

# plotting lines
ax.plot(line_df['mpg']/10, label='Mileage')
ax.plot(line_df['horsepower']/100, label='Horsepwoer')
ax.plot(line_df['weight']/1000, label='Weight')

# adding text in the plot
plt.text(76, 1.14, 'Horsepwer',
         va = 'center', # va-> vertical alignment
         rotation = 4, # angle
         bbox=dict(
            boxstyle = 'round',
            facecolor = 'wheat',
            alpha = 0.3
          )
         ) # background styles

plt.text(76,2.35, 'Mileage',
         va = 'center',
         rotation = 20,
         bbox = dict(
            boxstyle = 'round',
            facecolor = 'wheat',
            alpha = 0.3
          )
         )

plt.text(76, 3.15, 'Weight',
         va = 'center',
         rotation = -13,
         bbox = dict(
            boxstyle = 'round',
            facecolor = 'wheat',
            alpha = 0.3
          )
         )

# pointing to a specific position 'xy' using an arrow and
#   placing the descriptive text at position 'xytext'
plt.annotate('Highest Mileage', xy = (80.1, 3.39),
             xytext = (80.5, 3.38),
             arrowprops = dict(
                facecolor = 'wheat',
                shrink = .01
              ) # using arrow to point at 'xy'
            )

plt.annotate('Lowest Mileage', xy = (73.2, 1.70),
             xytext = (74, 1.66),
             arrowprops = dict(
                facecolor = 'wheat',
                shrink = .01
             )
            )

ax.set_title('Mileage vs Horsepower vs Weight', fontsize=15)
ax.set_xlabel('Model Year', fontsize=12)
ax.set_ylabel('Values', fontsize=12)

ax.legend()
fig.show()
```

# Markers

## Markers are used to highlight the intercepts in a plot.
## There are several types of markers such as 'o', 's', 'x', '+', etc.

### Using the marker paramter in a plot:

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[8,4])
ax = fig.add_axes([0,0,1,1])

# plotting the lines
ax.plot(line_df['mpg']/10, label = 'Mileage',
        color = "g",
        marker = '+',
        markersize = 7
        )
ax.plot(line_df['horsepower']/100, label = 'Horsepower',
        color = "b",
        marker = 's',
        markersize = 7
        )
ax.plot(line_df['weight']/1000, label = 'Weight',
        color = "r",
        marker = 'o',
        markersize = 7
        )

ax.set_title('Mileage vs Horsepower vs Weight', fontsize=15)
ax.set_xlabel('Model Year', fontsize=12)
ax.set_ylabel('Values', fontsize=12)

ax.legend()
fig.show()
```
