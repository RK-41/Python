# HISTOGRAM

## A histogram also represents data as rectangular bars. Unilike the bar graph, it is used for continuous data. Each bar groups the no. into intervals (bins) and the height of the bar is based on the no. of values that fall into the corresponding intervals.

## A histogram is idealy suited to obtain the frequency distribution of a given data.

## Prob. St.: The engineers at XYZ Custom Cars want to identify the distribution of horsepower.

## Sol.: This can be done with help of a histogram.

## Syntax:

> ax.hist(x,bins)

*   ax: represents axes
*   x: input values - single array or seq. of arrays
*   bins: int or seq or str. If bin is integer, it defines the no. of equal-width bins in the range.





### Creating a Histogram for Horsepower:

```python
# importing required packages
import matplotlib.pyplot as plt

# creating an empty canvas/figure
fig = plt.figure(figsize=[8,6])

# adding axes
ax = fig.add_axes([0,0,1,1])
ax.set_title('Distribution of Horsepower', fontsize=15)
ax.set_ylabel('Frequency', fontsize=13)
ax.set_xlabel('Horsepower', fontsize=13)

# plotting histogram
ax.hist(df['horsepower'], bins=20)
fig.show()
```
