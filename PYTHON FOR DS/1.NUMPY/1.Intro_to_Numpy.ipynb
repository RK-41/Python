# NUMPY OBJECT CREATION
## Numpy array can be created using array() fn.
## The array() fn in Numpy returns an array object named ndarray.

## Syntax: np.array(object,dtype)
###   object - A python obj eg, a list
###   dtype - data type of obj eg, integer

```python
import numpy as np

# creating 1-d array
student_marks_arr = np.array([34,64,69,54,26])
student_marks_arr
```

```python
# Creating a numpy array from python list

pylist = [23,34,46,28,76]
nparr = np.array(pylist)
nparr
```

```python
pylist
```

```python
# Shape of the array
nparr.shape
```

```python
# Data type of data contained by the array
nparr.dtype
```

```python
list_hetero = [23,9.2,"oi"]
arr_hetero = np.array(list_hetero)
arr_hetero
```

```python
arr_hetero.dtype
```

```python
# Changing dtype

list_int = [4,5,7,23]
arr_float = np.array(list_int, dtype='float')
arr_float
```

# WORKING WITH LARGE DATA

```python
# Calculating a million integers using Python lists

%%time
# used to calculate total operation time

list1=list(range(1,1000000))
list2=list(range(2,1000001))
list3=[]

for i in range(len(list1)):
  list3.append(list1[i]+list2[i])
```

```python
# Calculating a million integers using NumPy

%%time

import numpy as np

# creating numpy arrays of 1 million numbers
a = np.arange(1,1000000)
b = np.arange(2,1000001)
c = a+b
```

## A huge operational time gap can be observed:
  ### List: Wall time: 703 ms
  ### Numpy array: Wt: 24.4 ms
