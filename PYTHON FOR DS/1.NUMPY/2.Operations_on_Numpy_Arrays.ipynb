# ACCESSING ELEMENTS FROM A 2D ARRAY

```python
import numpy as np

# Creating a 2d array consisting of car names and horsepower
car_names = ['chevrolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino']
horsepower = [130, 165, 150, 150, 140]
car_hp_arr = np.array([car_names,horsepower])
car_hp_arr
```

```python
# Accessing car names
car_hp_arr[0]
```

```python
# Accessing 2nd car name
car_hp_arr[0,1]
```

```python
# Accessing last car name
car_hp_arr[0,-1]
```

# SLICING
## Slicing is a way to access and obtain subsets of ndarray in numpy

```python
# Slicing from 1D array

car_names[1:3]
```

```python
# Slicing from a 2D array
acceleration = [18, 15, 18, 16, 17]
car_hp_acc_arr = np.array([car_names,horsepower,acceleration])
car_hp_acc_arr
```

```python
car_hp_acc_arr[0:2]
```

```python
# Accessing name and horsepower of last two cars
car_hp_acc_arr[0:2,3:5]

# '0:2' gets car_names and horsepower
# '3:5' gets last two data from car_names and horsepower
```

```python
# Accessing name, hp, and acc of last car
car_hp_acc_arr[0:3,-1]
```

# MEAN AND MEDIAN

```python
# Creating a np array for HP
hp_arr = np.array(horsepower)
hp_arr
```

```python
# Mean of HP
print("Mean horsepower: ", np.mean(hp_arr))
```

```python
# Median of HP
print("Median horsepower: ", np.median(hp_arr))
```

# MIN & MAX

```python
# Min & Max HP
print("Minimum HP: ", np.min(hp_arr))
print("Maximum HP: ", np.max(hp_arr))
```

```python
# FINDING INDEX OF MINIMUM & MAXIMUM VALUES

print("Index of Min HP: ", np.argmin(hp_arr))
print("Index of Max HP: ", np.argmax(hp_arr))
```

```python
# FINDING HP OF CARS THAT ARE GREAER THAN OR EQUAL TO 150
#   Solution: 'where()' fn

x = np.where(hp_arr >= 150)
print("Indices of HP >= 150: ", x)

# with the indices, HP can be found
print("HP >= 150: ", hp_arr[x])
# hp_arr[x]
```

# CREATING A SEPARATE ARRAY CONSISTING OF FILTERED VALUES



```python
# Creating an array hp > 135

filter_arr = hp_arr > 135
# returns a boolean array for given condition

newarr = hp_arr[filter_arr]
# creates an array of elements satisfying the condition

print(filter_arr)
print(newarr)
```

# SORTING NUMPY ARRAY
## There are two ways to sort an array:
### Using array.sort() fn and numpy.sort(array) fn.
## array.sort() fn modifies the original array by default, whereas np.sort(array) fn does not.

```python
# Making a copy array of hp_arr
hp_copy_arr = hp_arr
```

```python
# Using sort(array) fn
print("Original array: ", hp_copy_arr)
print("Sorted array: ", np.sort(hp_copy_arr))
print("Original array after sorting: ", hp_copy_arr)
```

```python
# Using array.sort() fn
print("Original array: ", hp_copy_arr)
hp_copy_arr.sort()
print("Original array after sorting: ", hp_copy_arr)
```

# Numpy Arrays and Mathematical Operations
## Numpy makes use of optimized, pre-compiled code to perform mathematica operations on each array element.
## This eliminates the need of using loops, thereby enhancing the performance.
## This process is called VECTORIZATION.
## Numpy provides various mathematicl fns such as sum(), add(), sub(), log(), sin(), etc.

Screenshot 2023-08-30 192130.png




```python
# CREATING AN NUMPY ARRAY OF MARKS OF A STUDENT
marks_arr = np.array([89, 93, 68, 82, 59])
marks_arr
```

```python
# Calculating sum of marks
print(np.sum(marks_arr))
```

```python
# Awarding extra marks in every subjects
additional_marks = [2,3,5,1,8]
marks_arr += additional_marks
marks_arr
```

```python
# Another way to is to use add() fn
print("Original Marks: ", marks_arr)
new_marks = np.add(marks_arr,additional_marks)
print("New Marks: ", new_marks)
```

# BROADCASTING
## It's how Numpy handles arrays with different shapes during arithmetic operations.
## Array of smaller size is STRETCHED or COPIED across the larger array.

image.png

### Examples:

```python
# Ex.1:

arr1 = np.array([4,5,2])
arr2 = np.array([11])
arr3 = np.array([[1],[4],[8]])

print(arr1)
print(arr2)
print(arr3)

print("arr1+arr2: ", arr1+arr2)
print("arr1*arr2: ", arr1*arr2)
print("arr1+arr3: ", arr1+arr3)
```

```python
# Ex.2:
# A 2D array st_marks (4x2) consists of marks of 4 students in 2 subjects Py and C++

st_marks = np.array([[54,83],[89,75],[72,91],[64,82]])
print("Students marks: \n", st_marks)

# Incrementing marks in Py by 10 and in C++ by 5 using an 1D array
inc = np.array([10,5])

st_marks += inc
print("Marks after increment of [10,5]: \n", st_marks)
```
